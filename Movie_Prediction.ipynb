```python
TicketSale=pd.read_csv('/content/AnnualTicketSales.csv')
TicketSale.drop(TicketSale.columns[TicketSale.columns.str.contains("Unnamed",case=False)],inplace=True,axis=1)
TicketSale.head()
```
